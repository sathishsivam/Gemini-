```python
from transformers import BitsAndBytesConfig,AutoModelForCausalLM
```

```python
from google.colab import userdata
```

```python
import os
os.environ["HF_TOKEN"]=userdata.get("huggingface")
```

```python
import torch
```

```python
bit_config=BitsAndBytesConfig(load_in_4bit=True,bnb_4bit_quant_type="nf4",bnb_4bit_compute_dtype=torch.bfloat16)
```

```python
model=AutoModelForCausalLM.from_pretrained("EleutherAI/polyglot-ko-5.8b",device_map={"":0},quantization_config=bit_config)
```
